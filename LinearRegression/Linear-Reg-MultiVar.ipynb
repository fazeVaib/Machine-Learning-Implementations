###### Loading Boston Housing Data

```python
data = datasets.load_boston()
```

```python
df = pd.DataFrame(data.data, columns= data.feature_names)
# df
x = df.values
df
```

```python
y = pd.DataFrame(data.target, columns = ["MEDV"]).values
x = (x - x.mean())/x.std()
print(x[1], y)
```

```python
m,n = df.shape
theta = np.zeros([1,n+1])
print(theta)
```

```python
ones = np.ones([x.shape[0],1])
x = np.concatenate((ones,x),axis=1)
print(theta.shape)
print(x[1].shape)
print(theta[0])
```

###### Defining Hypothesis

```python
def hypo(x, theta):
    return x @ theta.T
```

######  Computing cost

```python
def computeCost(X, y, theta):
    tobesummed = np.power(((X @ theta.T) -y),2)
    return np.sum(tobesummed)/(2 * len(X))
```

###### Computing Gradient 

```python
def gradientDescent(X,y,theta,iters,alpha):
    cost = np.zeros(iters)
    for i in range(iters):
        theta = theta - (alpha/len(X)) * np.sum(X * (X @ theta.T - y), axis=0)
        cost[i] = computeCost(X, y, theta)
    
    return theta,cost
```

###### Continuing main function 

```python
alpha = 0.1
iterations = 20000
g,cost = gradientDescent(x,y,theta,iterations,alpha)
print(g)

finalCost = computeCost(x,y,g)
print(finalCost)
```

```python
plt.plot(np.arange(iterations), cost, 'g')  
plt.xlabel("Iterations")  
plt.ylabel('Cost')  
plt.show()
```

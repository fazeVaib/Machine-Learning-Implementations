## Support Vector Machine (SVM)

###### Importing Libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.io import loadmat
import seaborn as sb # for pairplots (not used here though xD) 
from sklearn import svm
%matplotlib inline
```

###### Loading data 

```python
data = loadmat('ex6data1.mat')
data['X']
# data
```

```python
mdata = pd.DataFrame(data['X'], columns=['X1', 'X2'])
mdata['y'] = data['y']
mdata.head()
```

###### Plotting data in graph 

```python
pos = mdata[mdata['y'].isin([1])]
neg = mdata[mdata['y'].isin([0])]

fig, ax = plt.subplots(figsize=(9,6))
ax.scatter(pos['X1'], pos['X2'], s=50, marker = 'o', alpha = 0.7, label = 'Positive', c='b')
ax.scatter(neg['X1'], neg['X2'], s=50, marker = 'x', alpha = 0.7, label = 'Negative', c='g')
ax.legend()
```

###### Implementing built-in SVM in scikit library 

```python
svc = svm.LinearSVC(C=1, loss='hinge', max_iter=1000)
# Linear Support Vector Classification, like SVC with linear kernel
svc
```

###### taking c=1, and checking the performance 

```python
svc.fit(mdata[['X1','X2']], mdata['y'])
svc.score(mdata[['X1', 'X2']], mdata['y'])
```

Since it is not 1.0, it seems to have miss-classified some value.
<br>
Trying out some other value may work (increasing c for more strict decision boundary)
<br>
But this may not result in a boundary that is normal fit for data!

```python
svc2 = svm.LinearSVC(C=100, loss='hinge', max_iter=1000)
svc2.fit(mdata[['X1','X2']], mdata['y'])
svc2.score(mdata[['X1', 'X2']], mdata['y'])
```

###### Plotting confidence level of the predicted data 

```python
mdata['SVM1 Confidence levels'] = svc.decision_function(mdata[['X1','X2']])

fig, ax = plt.subplots(figsize=(9,6))
ax.scatter(mdata['X1'], mdata['X2'], s=50, c=mdata['SVM1 Confidence levels'], cmap='seismic')
ax.set_title('SVM1 (C=1) Decision Confidence Levels')
```

Doing same for second SVM machine with high value of C

```python
mdata['SVM2 Confidence levels'] = svc2.decision_function(mdata[['X1','X2']])

fig, ax = plt.subplots(figsize=(9,6))
ax.scatter(mdata['X1'], mdata['X2'], s=50, c=mdata['SVM2 Confidence levels'], cmap='seismic')
ax.set_title('SVM2 (C=100) Decision Confidence Levels')
```

We can see, in second figure, the confidence levels of points near the boundary are pretty low, whereas in first one, the confidence level is pretty good.

### SVM using Gaussian Kernel (Non-linear)

```python
def gaussianKernel(x1, x2, sigma):  # WE can use built-in one too in scikit library
    return np.exp(-(np.sum((x1 - x2) ** 2) / (2 * (sigma ** 2))))

# trying out gaussian kernel
x1 = np.array([1.0, 2.0, 1.0])  
x2 = np.array([0.0, 4.0, -1.0])  
sigma = 2  
gaussianKernel(x1, x2, sigma)  
```

This is what value was expected to be. [Checked it from a course assignment in Coursera]

###### Now applying Gaussian Kernel on different Dataset with non-linear decision boundary 

```python
rawData = loadmat('ex6data2.mat')
data = pd.DataFrame(rawData['X'], columns=['X1', 'X2'])
data['y'] = rawData['y']
data.head()
```

Plotting the data on graph 

```python
pos = data[data['y'].isin([1])]
neg = data[data['y'].isin([0])]

fig, ax = plt.subplots(figsize=(9,6))
ax.scatter(pos['X1'], pos['X2'], s=50, marker = 'o', alpha = 0.7, label = 'Positive')
ax.scatter(neg['X1'], neg['X2'], s=50, marker = 'x', alpha = 0.7, label = 'Negative')
ax.legend()
```

Applying SVM on this data 
<br> <br>
Using Built-in RBF (Gausssian Kernel SVM) from the scikit library

```python
svc = svm.SVC(C=100, gamma=10, probability=True)
svc.fit(data[['X1', 'X2']], data['y'])
data['Probability'] = svc.predict_proba(data[['X1', 'X2']])

fig, ax = plt.subplots(figsize=(9,6))
ax.scatter(data['X1'], data['X2'], s=50, c = data['Probability'], cmap='Greens')
```

We could have used decision one too, but that was not so accurate in visualizing this type of data.
<br>
Hence, Instead we are using Probability prediction percent.

###### Applying SVM on validation set , after getting parameters from training set (3rd dataset)

```python
rdata = loadmat('ex6data3.mat')
# rdata
X = rdata['X']
Xcv = rdata['Xval'] # cross-validation data
y = rdata['y'].ravel() # flatten the data
ycv = rdata['yval'].ravel()
```

Defining values for C to test on and finding the best suited

```python
Cval = [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30, 100] # C values  
gval = [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30, 100] # Gamma values

bestScore = 0
bestParams = {'C': None, 'gamma': None}

for C in Cval:
    for gamma in gval:
        svc = svm.SVC(C=C, gamma=gamma)
        svc.fit(X, y)
        score = svc.score(Xcv, ycv)
        
        if score > bestScore:
            bestScore = score
            bestParams['C'] = C
            bestParams['gamma'] = gamma

bestScore, bestParams
```

#### Spam Classifier using  SVM on preprocessed data

```python
spamTrain = loadmat('spamTrain.mat')
spamTest = loadmat('spamTest.mat')

spamTrain
```

```python
X = spamTrain['X']
y = spamTrain['y'].ravel()
Xtest = spamTest['Xtest']
ytest = spamTest['ytest'].ravel()

X.shape, y.shape, Xtest.shape, ytest.shape
```

Here, 4000 are the training examples.
<br>
1899 are the parameters (i.e words defined in the dictionary for spam classifier)
<br>
So, X contains 0 or 1 value for each of the 1899 words described in dictionary and train through the given y value

###### Training the svm 

```python
svc = svm.SVC() # will take the default values for the parameters
svc.fit(X, y)

print('SVM model Accuracy on cross-validation: {}%'.format(np.round(svc.score(Xtest, ytest)*100, 5)))
```
